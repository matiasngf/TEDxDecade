```python
import pandas as pd
import requests
from IPython.display import HTML, display
from bs4 import BeautifulSoup
from bs4 import Comment
import json
import html
```

```python
url_proxy = 'https://free-proxy-list.net/anonymous-proxy.html'
response_pxy = requests.get(url_proxy)

pxySoup = BeautifulSoup(response_pxy.text, 'html.parser')
```

```python
rowsPxy = []
headPxy = []
for tablePart in pxySoup.find_all(attrs={'id':'proxylisttable'})[0].find_all(True, recursive=False):
    if(tablePart.name == 'thead'):
        for row in tablePart.find_all(True, recursive=False):
            headPxy = [tr.text for tr in row]
    if(tablePart.name == 'tbody'):
        for row in tablePart.find_all(True, recursive=False):
            rowsPxy.append([tr.text for tr in row])

proxiesDf = pd.DataFrame(data=rowsPxy, columns=headPxy)
proxiesDf = proxiesDf[(proxiesDf['Anonymity']=='elite proxy')].sort_values(by='Https', ascending=False)
proxiesDf.head()
```

```python
headersFirefox = {
    'User-Agent': 'Mozilla/5.0 (Android 4.4; Mobile; rv:41.0) Gecko/41.0 Firefox/41.0'
}
proxies = {
    "http": proxiesDf[proxiesDf.Https=='no'].loc[:,'IP Address'].values[0]
}
```

```python
responseYt = requests.get('https://m.youtube.com/user/TEDxTalks/videos?view=0&sort=dd&flow=list', headers=headersFirefox, proxies=proxies)
```

```python
ytData = json.loads(BeautifulSoup(responseYt.text, 'html.parser').find_all(attrs={'id':'initial-data'})[0].\
find_all(string=lambda text: isinstance(text, Comment))[0])
```

```python
rawVideos = pd.DataFrame(columns=['videoId', 'titleRaw', 'publishedTime', 'viewCount', 'length'])
def get_first_key(ytData):
    secondPageKey = False
    for i in ytData['contents']['singleColumnBrowseResultsRenderer']['tabs']:
        if('contents' in i['tabRenderer']['content']['sectionListRenderer']):
            for content in i['tabRenderer']['content']['sectionListRenderer']['contents']:
                for continuations in content['itemSectionRenderer']['continuations']:
                    secondPageKey = continuations['nextContinuationData']['continuation']
    return secondPageKey
userAgentFirefoxMobile = 'Mozilla/5.0 (Android 4.4; Mobile; rv:41.0) Gecko/41.0 Firefox/41.0'
headersReloadMore =  {
    'User-Agent': userAgentFirefoxMobile,
    'x-spf-referer': 'https://m.youtube.com/user/TEDxTalks/videos?view=0&sort=dd&flow=list',
    'X-YouTube-Client-Name': '2',
    'X-YouTube-Client-Version': '2.20190502',
}
def getContinuationKey(section):
    return section['continuations'][0]['nextContinuationData']['continuation']

def getVideoData(vcontent):
    title = vcontent['compactVideoRenderer']['title']['runs'][0]['text']
    thumbnails = vcontent['compactVideoRenderer']['thumbnail']['thumbnails']
    videoId = vcontent['compactVideoRenderer']['videoId']
    publishedTime = vcontent['compactVideoRenderer']['publishedTimeText']['runs'][0]['text']
    viewCount = vcontent['compactVideoRenderer']['viewCountText']['runs'][0]['text']
    length = vcontent['compactVideoRenderer']['lengthText']['runs'][0]['text']
    rawVideos.loc[videoId] = [videoId, title, publishedTime, viewCount, length]
    return [videoId, title, publishedTime, viewCount, length]
```

```python
for i in ytData['contents']['singleColumnBrowseResultsRenderer']['tabs']:
    if('contents' in i['tabRenderer']['content']['sectionListRenderer']):
        for vcontent in i['tabRenderer']['content']['sectionListRenderer']['contents'][0]['itemSectionRenderer']['contents']:
            getVideoData(vcontent)
            
firstKey = get_first_key(ytData)
print(firstKey)
```

```python
maxPage = 100
currPage = 0
def getNextSection(key):
    global currPage
    pageResponse = requests.get(
        'https://m.youtube.com/user/TEDxTalks/videos?pbj=2&ctoken='+key,
        headers=headersReloadMore, proxies=proxies
    )
    pageData = json.loads(pageResponse.text)
#     print(pageData['response']['continuationContents']['itemSectionContinuation']['contents'])
    nextKey = getContinuationKey(pageData['response']['continuationContents']['itemSectionContinuation'])
    for vcontent in pageData['response']['continuationContents']['itemSectionContinuation']['contents']:
        getVideoData(vcontent)
    print(currPage+1, nextKey)
    if(currPage < maxPage):
        currPage += 1
        getNextSection(nextKey)

getNextSection(firstKey)
```

```python
def cleanRaw(raw):
    titleEvent = raw.titleRaw.str.extract(r'(.+) \| ([^|]+) \| ([^|]+)$')
    titleEvent.columns = ['title', 'speaker', 'event']
#     display(raw)
#     print(int('12,345'))
    views = pd.DataFrame(
        raw.viewCount.str.replace(r'[^\d]', '').apply(pd.to_numeric),
    )
    views.columns = ['views']
#     display(raw.join(titleEvent).join(views))
    return raw.join(titleEvent).join(views)
TEDx = cleanRaw(rawVideos)
print(len(TEDx))
TEDx
```
