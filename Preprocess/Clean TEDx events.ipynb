```python
def clearShortName(name):
    if(re.search('@', name)):
        name = re.findall(r'@(.+)$', name)[0]
    for x in ['Salon','Youth','Live']:
        name = re.sub(x, '', name)
    return name

events = eventsDf.copy()
events.loc[:,'shortName'] = events.name.apply(lambda x: x[4:])
# events.loc[events[events.name.apply(lambda x: '@' in x)].index, 'shortName'] = events[events.name.apply(lambda x: '@' in x)].name.apply(lambda x: re.findall(r'@(.+)$', x)[0])

events.loc[:,'shortName'] = events.shortName.apply(clearShortName)
events[events.country=='Argentina'].name.unique()
events
```
